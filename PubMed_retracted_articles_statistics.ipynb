# Crossref API

```python
from crossref.restful import Works

works = Works()

w1 = works.filter(update_type = 'retraction')
w2 = works.filter(update_type = 'Retraction')
w3 = works.filter(update_type = 'retration')

doi_list = []
for i in w1:
    doi_list.append(i['DOI'])
    
for i in w2:
    doi_list.append(i['DOI'])
    
for i in w3:
    doi_list.append(i['DOI'])
```

```python
len(doi_list)
```

```python
crossref_retractions = list(set(doi_list))
```

```python
from crossref.restful import Works

works = Works()

pubmed_publication_type = []
count = 0

for i in pub_citations:
    if works.doi(i) is None:
        pubmed_publication_type.append('Not Available')
        count += 1
        print(count)
    else:
        pubmed_publication_type.append(works.doi(i)['publisher'])
        count += 1
        print(count)
```
